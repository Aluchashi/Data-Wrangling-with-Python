De Morgan Theorem
* (a) (A ∩ B)' = A' ∪ B'.
* (b) (A ∪ B)' = A' ∩ B'.

```python
# (a)

sobai = {"darshonik","utsuk","chinta","sukhi","dukhi","abul","kabul","dabul","babul"}
hati = {"darshonik","utsuk","chinta","sukhi","dukhi"}
manush = {"abul","kabul","dabul","babul", "dukhi", "darshonik"}
```

```python
lhs = sobai - (hati.intersection(manush))
```

```python
rhs = (sobai - hati).union(sobai-manush)
```

```python
lhs == rhs
```
