#### Random Walks

```python
import random
position = 0
walk = [position]
nsteps = 1000
for _ in range(nsteps):
    step = 1 if random.randint(0, 1) else -1
    position += step
    walk.append(position)
```

```python
npsteps = 1000
```

```python
rng = np.random.default_rng(seed=12345)
```

```python
draws = rng.integers(0,2,size=npsteps)
```

```python
steps = np.where(draws==0,1,-1)
```

```python
walk = steps.cumsum()
```

```python
walk.min()
```

```python
walk.max()
```

```python
np.abs(walk) >= 10
```

```python
(np.abs(walk) >= 10).argmax()
```
