```python
sns.set_style("ticks")
sns.set_context("notebook")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "species", palette = "Set1")

sns.despine()
```

```python
sns.scatterplot(data = penguins, x = "species", y= "body_mass_g", hue = "island", style = "sex", alpha = 0.5, palette = "Set1")
```

```python
sns.stripplot(data = penguins, x = "species", y= "body_mass_g", hue = "island", dodge = True, jitter=True, palette = "Set1")
```

```python
sns.swarmplot(data = penguins, x = "species", y= "body_mass_g", hue = "island")
```

```python
#regression plot:
```

```python
sns.regplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm")
```

```python
sns.lineplot(data = penguins, x = "body_mass_g", y="flipper_length_mm")
```

```python
#Joint Plot

sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm")
```

```python
sns.jointplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue = "sex")
```

```python
#bar Plot:

sns.barplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex", palette=["Pink","Skyblue"], estimator=np.sum)
```

```python
#Countplot:

sns.countplot(data = penguins, x = "species", hue = "island", palette="Set2")
```

```python
# Box Plot:

sns.boxplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex", palette="Set2")
```

```python
#voiolin plot:

sns.violinplot(data = penguins, x="species", y="body_mass_g", hue = "sex", palette="Set3")
```

```python
#KDE Plot:

sns.kdeplot(data = penguins,x = "body_mass_g", hue = "species", palette="Set2", fill = True)
```

```python
columns = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
```

```python
#Heatmap:

sns.heatmap(data = penguins[columns].corr(), annot = True, cmap = "YlOrRd", linewidths=2)
```

```python
#Rug Plot:

sns.rugplot(data = penguins, x ="body_mass_g", hue = "species", palette = "Set2")
```

```python
#Pair Plot:

sns.pairplot(data = penguins, hue = "sex", palette = "Set2", diag_kind="kde")
```

```python
# pair grid -> subplot grid for plotting pairwise relationship

graph = sns.PairGrid(data = penguins, hue = "sex", palette = "Set2")
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
```
