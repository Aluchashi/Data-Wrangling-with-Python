#### Pyplot API

```python
#Univariate - Numerical

hati ={
    "Points" : [50,56,67,72,68,88,98,112,120,117,234,123,76,87,112,45,65,52,71,74]
}
```

```python
df = pd.DataFrame(hati)
```

```python
df.head()
```

```python
df.shape
```

```python
#Line Plot

plt.plot(df["Points"], color = "orange", marker = "o", linewidth = 1)
plt.grid()
plt.show()
```

```python
# Histogram

plt.hist(df["Points"], color = "red", label = "Points", bins = 5)
plt.show()
```

```python
#Boxplot

plt.boxplot(df["Points"])
plt.show()
```

```python
#Univariate -> Categorical
```

```python
df["Team"] = ['Chinta', 'Sukhi', 'Dukhi', 'Utsuk', 'Chinta'] * 4
```

```python
df.head()
```

```python
df.info()
```

```python
#Pie Chart:

count = df["Team"].value_counts()
count
```

```python
plt.pie(count, labels=count.index, autopct = "%1.1f", explode = (0,0.1,0,0))
plt.axis('equal')
plt.show()
```

```python
#Countplot

plt.bar(count.index, count, color=["red", "white", "green", "orange"])
plt.show()
```

```python
print(df)
```

```python
chinta_points = df[df["Team"] == "Chinta"]["Points"]
chinta_points
```

```python
sukhi_points = df[df["Team"] == "Sukhi"]["Points"]
sukhi_points
```

```python
dukhi_points = df[df["Team"] == "Dukhi"]["Points"]
dukhi_points
```

```python
utsuk_points = df[df["Team"] == "Utsuk"]["Points"]
utsuk_points
```

```python
#Box plot:

plt.boxplot([chinta_points,sukhi_points, dukhi_points, utsuk_points])
plt.show()
```

```python
points_group = df.groupby("Team")["Points"].sum()
points_group
```

```python
plt.pie(points_group, labels = points_group.index, autopct = "%1.2f")
plt.axis("equal")
plt.show()
```

```python
 #bar plot:

chinta_mean = int(chinta_points.mean())
sukhi_mean = int(sukhi_points.mean())
dukhi_mean = int(dukhi_points.mean())
utsuk_mean = int(utsuk_points.mean())

plt.bar(["Chinta", "Sukhi", "Dukhi", "Utsuk"], [chinta_mean, sukhi_mean, dukhi_mean, utsuk_mean], color =["green", "orange", "red", "purple"])
plt.show()
```

```python
df["experience"] = [1,2,3,1.5,2,10,4,5,3,5,1.2,3,6,2,3,4,3,4,15,2]
```

```python
df.head()
```

```python
print(df)
```

```python
#Bubble plot

plt.scatter(df["Age"], df["Points"],s= df["experience"]*20, color = "skyblue", edgecolor = "blue")
plt.show()
```

```python
#2 numerical , 1 categorical.
```

```python
plt.scatter(df["Age"], df["Points"], c = df["Team"].map({"Chinta":"red",
                                                         "Dukhi":"orange",
                                                         "Sukhi":"skyblue",
                                                         "Utsuk":"white"}))
plt.legend(True)
plt.show()
```

```python
color = {"Chinta":"red", "Dukhi":"orange", "Sukhi":"skyblue","Utsuk":"white"}

for team, color in color.items():
    df_Team = df[df["Team"] == team]
    plt.scatter(df_Team["Age"], df_Team["Points"], c=color, label = team)

plt.legend()
```
